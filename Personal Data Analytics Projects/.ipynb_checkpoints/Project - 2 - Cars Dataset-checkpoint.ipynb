# Cars Dataset

```python
import pandas as pd
```

```python
data = pd.read_csv('./Data/2. Cars Data1.csv')
```

# Data Cleaning

- Find all Null Value in the dataset. If there are null values, then fill it the mean of that column.

```python
data.isnull().sum()
```

```python
# columns with object data type
object_cols = data.select_dtypes('object').columns

for col in object_cols:
    mode = data[col].mode()[0]
    
    data[col].fillna(mode, inplace = True)
    
# columns with numerical data type
numeric_cols = data.select_dtypes('float64').columns

for col in numeric_cols:
    mean = data[col].mean()
    
    data[col].fillna(mean, inplace = True)
```
